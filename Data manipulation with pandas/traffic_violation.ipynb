```python
import pandas as pd
import numpy as np
from scipy.stats import mode
```

```python
#loading dataset into dataframe#
df_traffic=pd.read_csv('C:\\Users\\Rohit\\Desktop\\ROhit\\Data Science with Python - Data Manipulation Problem Statement\\traffic_violations.csv',sep='\t',parse_dates=['Date_Of_Stop'],infer_datetime_format=True)
```

```python
print (df_traffic['Latitude'].mean())
```

```python
#filling missing values with mean value of integer/float columns and with mode values for string/object columns
df_traffic['Latitude'].fillna(np.nanmean(df_traffic['Latitude']),inplace=True)
df_traffic['Longitude'].fillna(np.nanmean(df_traffic['Latitude']),inplace=True)
df_traffic['Vehicle_Type'].fillna(mode(df_traffic["Vehicle_Type"].astype('str'), nan_policy='omit').mode[0], inplace=True)
df_traffic['Vehicle_Year'].fillna(mode(df_traffic["Vehicle_Year"].astype('str'), nan_policy='omit').mode[0], inplace=True)
```
